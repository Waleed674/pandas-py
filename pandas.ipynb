```python
import pandas as pd
students =pd.DataFrame({
    "student_id":[101,102,103,104,105,106,107,108],
    "name":["Ali","wahab","zoya","waleed","rayan","duad","eman","hassan"],
    "age":[20,21,23,23,21,20,19,20],
    "deparments":["se","cs","it","math","se","cs","it","math"],
    "marks":[89,67,98,55,77,71,89,49],
    "attendence":[77,88,80,90,85,75,83,91]})
print(students)
```

```python
def grade(marks):
    if marks >= 90:          
        return "A+"
    elif marks >= 80:        
        return "A"
    elif marks >= 70:        
        return "B"
    elif marks >= 60:        
        return "C"
    elif marks >= 50:        
        return "D"
    else:             
        return "F"

students["grade"] = students["marks"].apply(grade)
print(students)
```

```python
students["pass"]=students["marks"].apply(
    lambda x:"pass" if x>=50 else "fail")
print(students)
```
